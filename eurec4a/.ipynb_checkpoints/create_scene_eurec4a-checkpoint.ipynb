```python
# Read the numerical model output, make some maps to check the conditions, crop the fields of interest in the areas
# of interest and save them in a netcdf file.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import geometry as gm
```

```python
# To plot maps.
crs = ccrs.PlateCarree()

def plot_background(ax):
    #ax.set_extent(extent_param)
    ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
    #ax.add_feature(cfeature.STATES, linewidth=0.5)
    ax.add_feature(cfeature.BORDERS, linewidth=0.5)
    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, alpha=0.5)
    gl.top_labels = False
    gl.right_labels = False
    gl.xlabel_style = {'fontsize': 14}
    gl.ylabel_style = {'fontsize': 14}
    return ax
```

```python
def crop_field(field,lon,lat,area):
    """
    Function to crop a 2d field over the area specified by the list
    area = [minlon,minlat,maxlon,maxlat]
    """
    valid_indices = np.argwhere((lon>area[0])&(lon<area[2])&(lat>area[1])&(lat<area[3]))
    x_min, y_min = valid_indices.min(axis=0)
    x_max, y_max = valid_indices.max(axis=0)
    return field[x_min:x_max+1, y_min:y_max+1]
```

```python
# Set the paths and the file names of the atmospheric model outputs.
path2atm = '/home/agostino/SST-impact_data/wrf/'
```

```python
# Set the instant of interest.
str_instant = '2020-02-03T12:00:00'
# Valid range: from 2020-02-02_00:00:00 to 2020-02-03_23:00:00.

atm_filename = path2atm + 'wrfout_d02_' + str_instant[:10] + '_' + str_instant[11:]
```

```python
# Read the output files.
ds_atm = xr.open_dataset(atm_filename)

# Select the atmospheric variables of interest.
lon_atm = ds_atm['XLONG'][0].values
lat_atm = ds_atm['XLAT'][0].values
lakemask = ds_atm['LAKEMASK'][0].values
landmask = ds_atm['LANDMASK'][0].values
sst_atm = ds_atm['SST'][0].values
u10_mod = ds_atm['U10'][0].values
v10_mod = ds_atm['V10'][0].values
cosalpha = ds_atm['COSALPHA'][0].values
sinalpha = ds_atm['SINALPHA'][0].values

seamask = 1 - lakemask - landmask
seamask_nan = np.where(seamask==1,seamask,np.nan)
u10 = u10_mod*cosalpha - v10_mod*sinalpha
v10 = u10_mod*sinalpha + v10_mod*cosalpha
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1, projection=crs)
plot_background(ax)
pc = plt.pcolormesh(lon_atm,lat_atm,(sst_atm-273.15)/seamask,vmin=25,vmax=28)
step = 15

plt.title(str_instant + ' ocean surface temperature\nfrom atmospheric model [C]', fontsize=16)
plt.xlabel('Longitude',fontsize=14)
plt.ylabel('Latitude',fontsize=14)
plt.colorbar(pc)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1, projection=crs)
plot_background(ax)
pc = plt.pcolormesh(lon_atm,lat_atm,np.sqrt(u10**2+v10**2)*seamask_nan)
step = 15

plt.title(str_instant + ' wind speed [m/s]', fontsize=16)
plt.xlabel('Longitude',fontsize=14)
plt.ylabel('Latitude',fontsize=14)
plt.colorbar(pc)
```

```python
# Compute the wind divergence.
div = gm.div_sphere(u10,v10,lon_atm,lat_atm)

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1, projection=crs)
plot_background(ax)
pc = plt.pcolormesh(lon_atm,lat_atm,div*seamask_nan,vmin=-1e-3,vmax=1e-3,cmap='coolwarm')
step = 15

plt.title(str_instant + ' wind divergence [1/s]', fontsize=16)
plt.xlabel('Longitude',fontsize=14)
plt.ylabel('Latitude',fontsize=14)
plt.colorbar(pc)
```

```python
# Crop the relevant areas and save the netcdf file.
minlon = -60; minlat = 5; maxlon = -50; maxlat = 18.5
area_name = 'domain'

area = [minlon,minlat,maxlon,maxlat]


# Atmospheric variables (the cropped lon and lat are the same as for the oceanic model).
lon_c = crop_field(lon_atm,lon_atm,lat_atm,area)
lat_c = crop_field(lat_atm,lon_atm,lat_atm,area)
wind10_div_c = crop_field(div*seamask_nan,lon_atm,lat_atm,area)
u10_c = crop_field(u10*seamask_nan,lon_atm,lat_atm,area)
v10_c = crop_field(v10*seamask_nan,lon_atm,lat_atm,area)
sst_c = crop_field(sst_atm*seamask_nan,lon_atm,lat_atm,area)

# Create a dataset.
ds = xr.Dataset(
    data_vars = dict(
        lon=(["x","y"],lon_c),    
        lat=(["x","y"],lat_c),    
        sst=(["x","y"],sst_c,dict(units='K')),    
        u10=(["x","y"],u10_c,dict(units='m/s')),    
        v10=(["x","y"],v10_c,dict(units='m/s')),    
        wind_div=(["x","y"],wind10_div_c,dict(units='1/s'))    
    ),
    attrs=dict(
        ref_time = str_instant
    )
)

# Save the netcdf.
strfile = 'eurec4a_'+str_instant+'_area_'+area_name+'.nc'
ds.to_netcdf(strfile)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1, projection=crs)
plot_background(ax)
pc = plt.pcolormesh(lon_c,lat_c,sst_c-273.15)
step = 20
qv = plt.quiver(lon_c[::step,::step],lat_c[::step,::step],u10_c[::step,::step],v10_c[::step,::step],
                color='w',scale=100)
#ct = plt.contour(lon_c,lat_c,np.sqrt(u10_c**2+v10_c**2),colors='r')
plt.title(str_instant + ' SST [C] and sfc wind', fontsize=16)
plt.xlabel('Longitude',fontsize=14)
plt.ylabel('Latitude',fontsize=14)
plt.colorbar(pc)
figure_name = 'eurec4a_'+str_instant+'_area_'+area_name+'_sst_and_sfc_wind.png'
plt.savefig(figure_name,bbox_inches='tight') 
```
